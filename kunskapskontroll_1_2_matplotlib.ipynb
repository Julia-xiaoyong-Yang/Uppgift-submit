### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# 1. (Stateful interface) Pyplot style, with funciton "plt.plot()", simple, similar with MATLAB

import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Example Pyplot style')
plt.show()

# 2. Object-Oriented Interface, recommended for complex or customized plots.

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

fig, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(x, y)
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.set_title('Example Object-Oriented Interface')
plt.show()
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# A figure is the top-level container or window that holds all elements of a plot or visualization. It can contain multiple axes or subplots. Think of it as the canvas on which you create your plots.
# An axes presents a single plot or subplot within a figure. A figure can have one or more axes. 
# An axis refers to the x-axis or y-axis of a plot. Each axes typically has two axes: the x-axis and the y-axis.
# An artist is a low-level object in Matplotlib that represents individual graphical elements in a plot, such as lines, text, patches, or images. Artists are responsible for rendering and styling specific elements of the plot.
# In summary, a figure is the overall container, axes are the individual plots or subplots within a figure, axes have x and y axes, and artists are the graphical elements that make up the plot. 
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import matplotlib.pyplot as plt
import numpy as np
# import matplotlib as mpl

x = np.linspace(-4, 4, 100) # Sample data
y = x**2

# plt.figure(figsize=(5, 2.7), layout = "constrained")
plt.plot(x, y, label = "y = x^2", color="green")
plt.plot(x, y/2, label = "y = (x^2)/2", color="red")
plt.title("pyplot approach")
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(x, x**2, label = "y = x^2", color="blue")
ax.set_xlabel("x lable")
ax.set_ylabel("y lable")
ax.set_title("object-oriented approach")
ax.legend()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for scatter plot
np.random.seed(45)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,2.7))
ax1.scatter(random_data_x, random_data_y, alpha=0.5, color="red", label="Random")
ax1.set_xlabel("x label")
ax1.set_ylabel("y label")
ax1.set_title("A Scatter Plot")
ax1.legend()

ax2.bar(names, values, color="blue", label="Bar")
ax2.set_xlabel("x label")
ax2.set_ylabel("y label")
ax2.set_title("A Bar Plot")
ax2.annotate('look grapes!', xy=(0,20), xytext=(1,20),arrowprops=dict(facecolor='Yellow', shrink=0.15))
ax2.legend()

plt.tight_layout()
plt.show()
```
