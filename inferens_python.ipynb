## Create two empty vectors, x1 and x2. Now do a for loop 100 times, where you for each iteration: 
## 1) Take the mean of 100 samples from $N(10, 5)$ and store it in x1. 
## 2) Take the mean of 100 samples from $N(10, 20)$ and store it in x2.

## So, your vectors x1 and x2 should contain 100 values. 

## From p.148 in the book, we know that the mean is an unbiased estimate of the fixed, but unknown $\mu$. The nice thing is that in our Monte Carlo Simulation we know the true $\mu$. Plot histograms of x1 and x2 and comment on the result, do we on "average" get the right mean? Is it easier or harder to estimate the mean when the variance is low/high?
```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(100)
x1 = []
x2 = []

# Number of interations
iterations = 100

# Generate data and calculate means
for i in range(iterations):
    # Generate 100 random samples from N(10,5)
    samples1 = np.random.normal(10,5,100)
    # Calculate and store the mean
    x1.append(np.mean(samples1))
    
    # Generate 100 random samples from N(10,20)
    samples2 = np.random.normal(10, 20, 100)
    # Calculate and store the mean
    x2.append(np.mean(samples2))
    
# Plot histograms
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.hist(x1, bins=20, color='blue', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of x1 with N(10,5)')

plt.subplot(1,2,2)
plt.hist(x2, bins=20, color='red', alpha=0.7)
plt.title('Histogram of x2 with N(10,20)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

# Chapter 7 - Confidence Intervals

Assuming $X$ is $N(10, 3)$, we know (see example calculation below) that a random sample x will with probability 95% be between: <br>
ci_lower = 10 - 1.96 x 3 <br> and <br>
ci_upper = 10 + 1.96 x 3

Example, for the ci_upper, we do the following calculation: $P(X > ci\_upper) = P(X > 10 + 1.96 * 3) = P(Z > 1.96) = 0.025$ where we used the usual standardization. 

This means that if you get observations from the normal distribution above, you would "expect" 95 \% of them to be between ci_lower and ci_upper. 

Your task is now to simulate 10, 100, 1000 and 10000 normally distributed samples from $N(10, 3)$ and check the relative frequency (\%) of how many samples fall outside the confidence interval. How many \% should it theoretically be?

```python
ci_lower = 10 - 1.96*3
ci_upper = 10 + 1.96*3

print(ci_lower)
print(ci_upper)
```

```python
import numpy as np

# Define parameters
mean = 10
std_dev = 3
confidence_level = 0.95
sample_sizes = [10, 100, 1000, 10000]
num_simulations = 1000 # Number of simulations for each sample size

# Calculate confidence interval bounds

z_score = abs(np.percentile(np.random.normal(0, 1, 100000), (1-confidence_level)/2))
ci_lower = mean - z_score * (std_dev / np.sqrt(sample_sizes))
ci_upper = mean + z_score * (std_dev / np.sqrt(sample_sizes))

# Initialize dictionaries to store results
results = {n: [] for n in sample_sizes}

# Perform simulations
for sample_size in sample_sizes:
    for _ in range(num_simulations):
        # Generate random samples from N(10, 3)
        samples = np.random.normal(mean, std_dev, sample_size)
        
        # Calculate the percentage of samples outside the confidence interval
        outside_ci = np.logical_or(samples < ci_lower[sample_sizes.index(sample_size)], samples > ci_upper[sample_sizes.index(sample_size)])
        percentage_outside = np.mean(outside_ci) * 100
        
        # Append the result to the corresponding same size list
        results[sample_size].append(percentage_outside)
        
# Print result
for sample_size, percentages in results.items():
            print(f'Sample Size: {sample_size}')
            print(f'Mean Percentage Outside CI:{np.mean(percentages):.2f}%')
            print()
```

## Assume you have 100 observations from N(10, 3), you calculate a confidence interval and check if it "captures" the real $\mu$ (we know the reality since we are doing a monte carlo simulation), how often do you "capture" the real mu (in \%) when you repeat this 1000 times? How is this knowledge useful in reality?

## Note, the difference between this question and the one above is that here we calculate the sample mean and sample standard deviation to construct confidence intervals whereas above we used that we "knew the reality" ($\mu$ and $\sigma$), i.e. ci_lower <- 10 - 1.96*3 and  ci_upper <- 10 + 1.96*3. Obviosuly, this is not the case in reality.


```python
import numpy as np

# Define parameters
true_mean = 10
true_std_dev = 3
sample_size = 100
confidence_level = 0.95
num_simulations = 1000

# Initialize a counter to keep track of how ofter the trun mean is captured
capture_count = 0

# Perform the Monte Carlo simulaiton
for _ in range(num_simulations):
    # Generate s random sample of size 100 from N(10, 3)
    sample = np.random.normal(true_mean, true_std_dev, sample_size)
    
    # Calculate sample mean and sample standard deviation
    sample_mean = np.mean(sample)
    sample_std_dev = np.std(sample, ddof=1) # Use Bessel's correction (ddof = 1) for sample standard deviation
    
    # Calculate confidence interval bounds
    # "np.random.normal(0, 1, 100000)" generates 100,000 random numbers from a standard normal distribution (mean = 0, standard deviation = 1)
    # "(1 - confidence_level) / 2" calculates the tail probability for a two-tailed hypothesis test. For example, for a 95% confidence interval, this value would be (1 - 0.95) / 2 = 0.025, which corresponds to the 2.5th percentile of a standard normal distribution in each tail.
    # "np.percentile()" calculates the percentile of the data
    # abs() takes the absolute value
    z_score = abs(np.percentile(np.random.normal(0, 1, 100000), (1 - confidence_level)/2))
    ci_lower = sample_mean - z_score * (sample_std_dev / np.sqrt(sample_size))
    ci_upper = sample_mean + z_score * (sample_std_dev / np.sqrt(sample_size))
    
    # Check if the true mean is within the confidence interval
    if ci_lower <= true_mean <= ci_upper:
        capture_count += 1

# Calculate the percentage of times the true mean is captured
capture_percentage = (capture_count / num_simulations) * 100

print(f'Percentage of times the true mean is captured: {capture_percentage:.2f}%')
```

## Assume we ask 1000 people who they will vote for and 4.2% answer Centerpartiet (C). Create a 90% Confidence Interval to get the proportion that would vote for C if we surveyed the whole population ("Totalundersökning"). OBS: note we want a 90% confidence interval.

```python
import scipy.stats as stats

## Find the z-score for the 95th percentile of a standard normal distribution
# Calculates the z-score that corresponds to the 95th percentile of the standard normal distribution.
z_score = stats.norm.ppf(0.95)
print('Z_score for the 95th percentile:', z_score)
```

```python
import scipy.stats as stats
import math

# Define parameters
sample_proportion = 0.042 # 4.2%
simple_size = 1000
confidence_level = 0.90

# Calculate the critical value (Z) for a 90% confidence interval
z_critical = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the standard error
standard_error = math.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Calculate the margin of error
margin_of_error = z_critical * standard_error

# Calculate the confidence interval bounds
ci_lower = sample_proportion - margin_of_error
ci_upper = sample_proportion + margin_of_error

# Print the confidence interval
print(f'90% Confidence Interval for Proportion: ({ci_lower:.4f}, {ci_upper:.4f})')


```

## For this question see (p. 177 - 178) in the book. We ask 1000 people who they will vote for, 4.2% answer Centerpartiet (C) and 4.5% answer Liberalerna (L), is there a statistical significant difference between the two parties? 

## Check if the assumption on  np(1-p) > 5 is satisfied for both poportions.

```python
import numpy as np
import scipy.stats as stats

# Given values
n1 = 1000 # Sample size for C
n2 = 1000 # Sample size for L
p1 = 0.042 # Proportion for C 4.2%
p2 = 0.045 # Proportion for L 4.5%

# Calculate the standard errors for both proportions
set1 = np.sqrt((p1 * (1 - p1)) / 2)
set2 = np.sqrt((p2 * (1 - p2)) / 2)

# Calculate the standard error of the difference between proportions
se_diff = np.sqrt((set1**2 / n1) + (set2**2 / n2))

# Calculate the z-statistic for the difference between proportions
z_statistic = (p1 - p2) / se_diff

# Calculate the two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))

# Set the significance level
alpha = 0.05

# Check if the p-value is less than the significance level
if p_value < alpha:
    print('There is a statistically significant difference between C and L.')
else:
    print('There is no statistically significant difference between C and L.')
```
